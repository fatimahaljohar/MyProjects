This is accomplished by doing a convolution between a kernel and an image.filter2D(src, dst, ddepth, kernel)

* src âˆ’ A Mat object representing the source (input image) for this operation.

* ddepth âˆ’ A variable of the type integer representing the depth of the output image.

2. Opening is smoothes the contour of an object, breaks narrow isthmuses, and eliminates thin protrusions. Closing also tends to smooth sections of contours, but, as opposed to opening, it generally fuses narrow

* Morphological opening involves the application of erosion, followed by dilation as shown in the following equation <br>

$ A \circ B = (A \circleddash B) \oplus B $

```python
# Read image
img1 = cv2.imread('lady.JPG')
# Convert the image to colors-scale
original_img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
# Convert the image to gray-scale
gray_img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY) 

kernel = np.array([
  [0, -1, 0],
  [-1, 5, -1],
  [0, -1, 0]
])

# Apply convolution between image and 3x3 Kernel using cv2.filter2D
sharp1 = cv2.filter2D(gray_img1,-1, kernel) 

element = cv2.getStructuringElement(shape = cv2.MORPH_CROSS, ksize=(5,5))
opening = cv2.morphologyEx(sharp1, cv2.MORPH_OPEN, element)

opening = cv2.filter2D(opening,-1, kernel)
# Create a 3x3 Kernel
 
plt.figure(figsize=(10,5))
plt.subplot(121),plot_image(original_img1, 'Original image')
plt.subplot(122),plot_image(opening, 'Result')
```

```python
img5 = cv2.imread('poladrioddd.jpg')
original_img5 = cv2.cvtColor(img5,cv2.COLOR_BGR2RGB)
gray_img5 = cv2.cvtColor(img5,cv2.COLOR_BGR2GRAY)

# CLAHE is a variant of Adaptive histogram equalization (AHE) which takes care of over-amplification of the contrast. 
clahe = cv2.createCLAHE(clipLimit = 4) 
clahee = clahe.apply(gray_img5)

gamma = .8
gamma_imge = np.array(255*(clahee / 255) ** gamma, dtype = np.uint8)

kernel = np.array([
  [0, -1, 0],
  [-1, 5, -1],
  [0, -1, 0]
])
sharp = cv2.filter2D(gamma_imge,-1, kernel)
plt.figure(figsize=(10,5))
plt.subplot(141),plot_image(original_img5, 'Original image')
plt.subplot(142),plot_image(clahee, 'Equlaized')
plt.subplot(143),plot_image(gamma_imge, 'Gamma')
plt.subplot(144),plot_image(sharp, 'Result')
```

## Problem 6

Problem Definition : this image needed more clarity and edge enhancement <br>Approach description: We started to sharpen the image, since the first thing we noticed was that the image was slightly blurred, so we decided to apply sharpening. After checking the sharpening result, we found out that we needed to sharpen it again in order to enhance the details in it and it resulted with the second result. After the double sharpening, we found that we got a lot of noise, so we decided to use median blur in order to reduce salt and perpper noise giving us the final result.

```python
img6 = cv2.imread('baby.JPG') 
original_img6 = cv2.cvtColor(img6,cv2.COLOR_BGR2RGB)
gray_img6 = cv2.cvtColor(img6,cv2.COLOR_BGR2GRAY)

kernel = np.array([
  [0, -1, 0],
  [-1, 5, -1],
  [0, -1, 0]
])

# Apply convolution between image and 3x3 Kernel using cv2.filter2D
sharp = cv2.filter2D(gray_img6,-1, kernel)
sharp2 = cv2.filter2D(sharp,-1, kernel)
medianBlur = cv2.medianBlur(sharp2,3)
plt.figure(figsize=(10,5))
plt.subplot(141), plot_image(original_img6, 'Original image')
plt.subplot(142), plot_image(sharp,'sharp')
plt.subplot(143), plot_image(sharp2,'double sharp')
plt.subplot(144), plot_image(medianBlur,'Result')
plt.show()
```

## Problem 7 

Problem Definition : the colored image looks dull and needs some color and enhancements regarding clarity.<br> Approach description : We started with removing the salt and pepper noise by using median blur which resulted in the second image, then we needed to do some lighting and color adjustments using gamma, which gave us, a good results in the third image and finally we needed to sharpen the image to get a better sense of details in the final result.

```python
img7=cv2.imread('family.JPG')
original_img7 = cv2.cvtColor(img7,cv2.COLOR_BGR2RGB)

medain=cv2.medianBlur(original_img7,3)
#Adjust Gamma
gamma = 2
#Apply Gamma
img_gamma7 = np.array(255*(medain / 255) ** gamma, dtype = 'uint8')
kernel = np.array([
  [0, -1, 0],
  [-1, 5, -1],
  [0, -1, 0]
])

# Apply convolution between image and 3x3 Kernel using cv2.filter2D
sharp7= cv2.filter2D(img_gamma7,-1, kernel) 
plt.figure(figsize=(10,5))
plt.subplot(141), plot_image(original_img7, 'Original image')
plt.subplot(142), plot_image(medain,'Result')
plt.subplot(143), plot_image(img_gamma7,'Result')
plt.subplot(144), plot_image(sharp7,'Result')
plt.show()
```
